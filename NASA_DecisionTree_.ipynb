```python
print(df.shape) #Checks the number of rows and columns in dataframe.
df= df.dropna() #Removes the rows that contains NULL values.
print(df.shape) #Looks like there was some NULL values because the dropna() removes the rows
```

```python
def KM_to_AU(km):
  return km/149597870.7
```

```python
df['Est Dia in AU(min)'] = df['Est Dia in KM(min)'].apply(KM_to_AU)
```

### **Choosing and training the model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split #train_test_split divides our data into train and test sets
from sklearn.metrics import accuracy_score
```

```python
def avgScore(size, n):
  X = df[['Est Dia in AU(min)', 'Absolute Magnitude']]
  y = df['Hazardous']

  sum = 0

  for i in range(n):
    X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = size)

    model = DecisionTreeClassifier()

    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    score = accuracy_score(y_test, predictions)

    sum += score


  avg = sum/n
  return avg



print(avgScore(0.1, 10))
print(avgScore(0.2, 10))
print(avgScore(0.3, 10))
print(avgScore(0.4, 10))
print(avgScore(0.5, 10))
print(avgScore(0.6, 10))
print(avgScore(0.7, 10))
print(avgScore(0.8, 10))
print(avgScore(0.9, 10))

#test_size = 0.1 seems to produce the highest average accuracy_score. Therefore, we will use test_size = 0.1.
```

```python
X = df[['Est Dia in AU(min)', 'Absolute Magnitude']]
y = df['Hazardous']

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.1)

model = DecisionTreeClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

score = accuracy_score(y_test, predictions)

score
```
